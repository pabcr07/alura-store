### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)


tienda.head()
```

#1. Análisis de facturación



```python
total_tienda_1 = tienda["Precio"].sum()
total_tienda_2 = tienda2["Precio"].sum()
total_tienda_3 = tienda3["Precio"].sum()
total_tienda_4 = tienda4["Precio"].sum()
```

# 2. Ventas por categoría

```python
ventas_tienda_1 = tienda.groupby('Categoría del Producto').size()
ventas_tienda_1 = ventas_tienda_1.sort_values(ascending = False)

ventas_tienda_2 = tienda2.groupby('Categoría del Producto').size()
ventas_tienda_2 = ventas_tienda_2.sort_values(ascending = False)

ventas_tienda_3 = tienda3.groupby('Categoría del Producto').size()
ventas_tienda_3 = ventas_tienda_3.sort_values(ascending = False)

ventas_tienda_4 = tienda4.groupby('Categoría del Producto').size()
ventas_tienda_4 = ventas_tienda_4.sort_values(ascending = False)

```

# 3. Calificación promedio de la tienda


```python
promedio_tienda_1 = tienda["Calificación"].mean().round(2)
promedio_tienda_2 = tienda2["Calificación"].mean().round(2)
promedio_tienda_3 = tienda3["Calificación"].mean().round(2)
promedio_tienda_4 = tienda4["Calificación"].mean().round(2)
```

# 4. Productos más y menos vendidos

```python
productos_tienda_1 = tienda.groupby('Producto').size().sort_values(ascending= False)
max_ventas_tienda1 = productos_tienda_1.max()
min_ventas_tienda1 = productos_tienda_1.min()

mas_vendidos_tienda_1 = productos_tienda_1[productos_tienda_1 == max_ventas_tienda1]
menos_vendidos_tienda_1 = productos_tienda_1[productos_tienda_1 == min_ventas_tienda1]
```

```python
productos_tienda_2 = tienda2.groupby('Producto').size().sort_values(ascending= False)
max_ventas_tienda2 = productos_tienda_2.max()
min_ventas_tienda2 = productos_tienda_2.min()

mas_vendidos_tienda_2 = productos_tienda_2[productos_tienda_2 == max_ventas_tienda2]
menos_vendidos_tienda_2 = productos_tienda_2[productos_tienda_2 == min_ventas_tienda2]
```

```python
productos_tienda_3 = tienda3.groupby('Producto').size().sort_values(ascending= False)
max_ventas_tienda3 = productos_tienda_3.max()
min_ventas_tienda3 = productos_tienda_3.min()

mas_vendidos_tienda_3 = productos_tienda_3[productos_tienda_3 == max_ventas_tienda3]
menos_vendidos_tienda_3 = productos_tienda_3[productos_tienda_3 == min_ventas_tienda3]
```

```python
productos_tienda_4 = tienda4.groupby('Producto').size().sort_values(ascending= False)
max_ventas_tienda4 = productos_tienda_4.max()
min_ventas_tienda4 = productos_tienda_4.min()

mas_vendidos_tienda_4 = productos_tienda_4[productos_tienda_4 == max_ventas_tienda4]
menos_vendidos_tienda_4 = productos_tienda_4[productos_tienda_4 == min_ventas_tienda4]
```

# 5. Envío promedio por tienda

```python
envio_tienda_1 = tienda["Costo de envío"].mean().round(2)
envio_tienda_2 = tienda2["Costo de envío"].mean().round(2)
envio_tienda_3 = tienda3["Costo de envío"].mean().round(2)
envio_tienda_4 = tienda4["Costo de envío"].mean().round(2)
```



# 6.Visualización

```python
#GRAFICO 1
nombre_tiendas = ["Tienda 1","Tienda 2","Tienda 3","Tienda 4"]
valores_ingresos = [total_tienda_1, total_tienda_2, total_tienda_3, total_tienda_4]
plt.figure(figsize = (8,6))
plt.bar(nombre_tiendas, valores_ingresos, color=["blue","green", "red","orange"])
plt.title('Ingresos Totales por tienda')
plt.show()
```

```python
#GRAFICO 2
plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
plt.pie(ventas_tienda_1, labels=ventas_tienda_1.index, autopct='%1.1f%%')
plt.title('Tienda 1')

plt.subplot(2, 2, 2)
plt.pie(ventas_tienda_2, labels=ventas_tienda_2.index, autopct='%1.1f%%')
plt.title('Tienda 2')


plt.subplot(2, 2, 3)
plt.pie(ventas_tienda_3, labels=ventas_tienda_3.index, autopct='%1.1f%%')
plt.title('Tienda 3')


plt.subplot(2, 2, 4)
plt.pie(ventas_tienda_4, labels=ventas_tienda_4.index, autopct='%1.1f%%')
plt.title('Tienda 4')


plt.show()
```

```python
#GRAFICO 3
nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
promedios = [promedio_tienda_1, promedio_tienda_2, promedio_tienda_3, promedio_tienda_4]


plt.figure(figsize=(8, 6))
barras = plt.barh(nombres, promedios, color='blue')
plt.title('Calificacion promedio de clientes')

plt.show()
```

```python
#GRAFICO 4

datos_envio = [tienda["Costo de envío"],tienda2["Costo de envío"],tienda3["Costo de envío"],tienda4["Costo de envío"]]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']


plt.figure(figsize=(10, 6))
plt.boxplot(datos_envio, labels=nombres_tiendas)
plt.title('Distribucion de Costos de Envio por tienda')

plt.show()
```
